```python
#####################################
# Usage : calc the area of a house
# Author: yuanyuefeng
# Date  : 2017/04/14
#####################################
def polyArea(pts):
    s=0
    _len=len(pts)
    x,y=zip(*pts)
    j=_len-1
    for i in range(_len):
        s=s+(x[j]+x[i])*(y[j]-y[i])
        j=i
    return abs(s/2.)
    
def calcArea(deltaData):
    x = y = 0
    pts = []
    pts.append((x,y))
    isX = True
    for delta in deltaData:
        if isX:
            x += delta
            pts.append((x,y))
        else:
            y += delta
            pts.append((x,y))
        isX = not isX

    if abs(x) > 1e-5:
        print("Error！The sum of 'x' = %f, should be 0." % (x,))
    elif abs(y) > 1e-5:
        print("Error！The sum of 'y' = %f, should be 0." % (y,))
    elif not isX:
        print("Error！The data pair of (x,y) is in-complete.")
    else:
        print("Total Area: %.2f" % (polyArea(pts),))

calcArea(
   (#deltaX,  deltaY
    1.4,      1.6,
    2.1,      1.4,
    3,        1.2,
    2.2+1.9,  -0.6,
    1.8,      -(1.4+2.2+1.3+2.5+2.5),
    -3.7,     1.6,
    -2.8,     0.6,
    -2.5-3.4, 4.1)
)
```

```python
#####################################
# Usage : solve move one match issue
# Author: yuanyuefeng
# Date  : 2019/06/18
#####################################
transfer_table = {
    '0' : ( ('-0',1), ('8',1), ('6',0), ('9',0) ),
    '1' : ( ('-1',1), ('/1',1), ('7',1), ('1',-1) ),                      #特殊变换：('1',-1)
    '2' : ( ('-2',1), ('/2',1), ('3',0) ),
    '3' : ( ('-3',1), ('/3',1), ('9',1), ('2',0), ('5',0) ),
    '4' : ( ('-4',1), ('/4',1), ('-1',-1) ),                              #特殊变换：('-1',-1)
    '5' : ( ('-5',1), ('/5',1), ('6',1), ('9',1), ('3',0) ),
    '6' : ( ('-6',1), ('/6',1), ('8',1), ('5',-1), ('9',0), ('0',0) ),
    '7' : ( ('-7',1), ('/7',1), ('1',-1) ),
    '8' : ( ('-8',1), ('/8',1), ('0',-1), ('6',-1), ('9',-1) ),
    '9' : ( ('-9',1), ('/9',1), ('8',1), ('3',-1), ('5',-1), ('6',0), ('0',0) ),
    '+' : ( ('-',-1), ('/+',1), ('=',0) ),
    '-' : ( ('+',1), ('/-',1), ('=',1) ),
    'x' : ( ('/', -1), ),
    '/' : ( ('x', 1), ),
    '=' : ( ('-',-1), ('+',0) )
}

#Return code:
#1:correct，-1:correct if upside down, 0:wrong
def evalMatchEq(s):
    if s.count("=") == 1:
        new_s = s.replace('=', "==").replace('x', '*')
        try:
            if eval(new_s) == True:
                return 1
            else:
                charList = list(new_s[::-1])
                for i in range(len(charList)):
                    if charList[i] == '6':
                        charList[i] = '9'
                    elif charList[i] == '9':
                        charList[i] = '6'
                    elif charList[i] == '3' or \
                         charList[i] == '4' or \
                         charList[i] == '7' or \
                         charList[i] == '/':
                        return 0
                new_s = ''.join(charList)
                if eval(new_s) == True:
                    return -1
        except:
            pass
    return 0

def solveMoveOneMatch(s):
    solved = False
    #try any 0
    for i in range(len(s)):
        for new_c,match in transfer_table[s[i]]:
            if match == 0:
                charList = list(s)
                charList[i] = new_c
                new_s = ''.join(charList)
                ret = evalMatchEq(new_s)
                if ret == 1:
                    print("%s => %s" %(s, new_s))
                    solved = True
                elif ret == -1:
                    print("%s => %s(倒着看)" %(s, new_s))
                    solved = True
    
    #try any -1 & +1
    for i in range(len(s)):
        for new_c_i,match_i in transfer_table[s[i]]:
            if match_i == -1:
                charList_i = list(s)
                charList_i[i] = new_c_i
                new_s_i =  ''.join(charList_i)
                for j in range(len(new_s_i)):
                    for new_c_j,match_j in transfer_table[new_s_i[j]]:
                        if match_j == 1:
                            charList = list(new_s_i)
                            charList[j] = new_c_j
                            new_s = ''.join(charList)
                            ret = evalMatchEq(new_s)
                            if ret == 1:
                                print("%s => %s" %(s, new_s))
                                solved = True
                            elif ret == -1:
                                print("%s => %s(upside down)" %(s, new_s))
                                solved = True
    if not solved:
        print("%s Unsolvable" % (s,))

solveMoveOneMatch("12+7=10")
solveMoveOneMatch("53+9=67")
solveMoveOneMatch("18+18=8")
solveMoveOneMatch("7-6=5")
solveMoveOneMatch("4816=9")
solveMoveOneMatch("1+8=1")
solveMoveOneMatch("3+5=74")
solveMoveOneMatch("8-7=7")
solveMoveOneMatch("0+7=6")    #8+1=6(upside down)
solveMoveOneMatch("16+5=1")
solveMoveOneMatch("44+2=7")  #4+1+2=7
solveMoveOneMatch("3+0=8")
solveMoveOneMatch("5+5=55")
solveMoveOneMatch("2+8=7")
solveMoveOneMatch("13+12=7")
solveMoveOneMatch("2+6=11")
solveMoveOneMatch("2/3=8")
solveMoveOneMatch("7-6=5")
solveMoveOneMatch("7+1=97")
solveMoveOneMatch("19-3=6")
```
